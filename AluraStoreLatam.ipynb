### Importación de datos



```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#ingreso total de cada tienda sumando los valores de la columna precio de cada conjunto de datos

tienda_total = tienda["Precio"].sum()
tienda_total2 = tienda2["Precio"].sum()
tienda_total3 = tienda3["Precio"].sum()
tienda_total4 = tienda4["Precio"].sum()

print(f"{tienda_total:,.2f}")
print(f"{tienda_total2:,.2f}")
print(f"{tienda_total3:,.2f}")
print(f"{tienda_total4:,.2f}")
```

# 2. Ventas por categoría

```python
categoria_popular = tienda["Categoría del Producto"].value_counts()
categoria_popular2 = tienda2["Categoría del Producto"].value_counts()
categoria_popular3 = tienda3["Categoría del Producto"].value_counts()
categoria_popular4 = tienda4["Categoría del Producto"].value_counts()

print(categoria_popular,"\n")

print(categoria_popular2, "\n")

print(categoria_popular3, "\n")

print(categoria_popular4, "\n")
```

# 3. Calificación promedio de la tienda


```python
promedio_calificacion_tienda1 = (tienda["Calificación"].sum())/ len(tienda["Calificación"])
promedio_calificacion_tienda2 = (tienda2["Calificación"].sum())/ len(tienda2["Calificación"])
promedio_calificacion_tienda3 = (tienda3["Calificación"].sum())/ len(tienda3["Calificación"])
promedio_calificacion_tienda4 = (tienda4["Calificación"].sum())/ len(tienda4["Calificación"])


print(promedio_calificacion_tienda1)
print(promedio_calificacion_tienda2)
print(promedio_calificacion_tienda3)
print(promedio_calificacion_tienda4)
```

# 4. Productos más y menos vendidos

```python
productos_mas_vendidos = tienda["Producto"].value_counts()
productos_mas_vendidos2 = tienda2["Producto"].value_counts()
productos_mas_vendidos3 = tienda3["Producto"].value_counts()
productos_mas_vendidos4 = tienda4["Producto"].value_counts()

print(productos_mas_vendidos)
print(productos_mas_vendidos2)
print(productos_mas_vendidos3)
print(productos_mas_vendidos4)
```

# 5. Envío promedio por tienda

```python
envio_promedio = (tienda["Costo de envío"].sum())/len(tienda["Costo de envío"])
envio_promedio2 = (tienda2["Costo de envío"].sum())/len(tienda["Costo de envío"])
envio_promedio3 = (tienda3["Costo de envío"].sum())/len(tienda["Costo de envío"])
envio_promedio4 = (tienda4["Costo de envío"].sum())/len(tienda["Costo de envío"])


print(envio_promedio)
print(envio_promedio2)
print(envio_promedio3)
print(envio_promedio4)
```

```python
import matplotlib.pyplot as plt

for tienda in [tienda]:
  producto_duplicados = tienda.duplicated(subset='Producto')
  tienda_sin_duplicados = tienda[~producto_duplicados]

plt.plot(tienda_sin_duplicados["Precio"], tienda_sin_duplicados['Calificación'], 'ro')
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.title("Relación entre calificación y precio tienda 1")
plt.show()

mayor = tienda["Precio"].max()
menor = tienda["Precio"].min()

mayor2 = tienda_sin_duplicados["Precio"].max()
menor2 = tienda_sin_duplicados["Precio"].min()

print(mayor)
print(menor)

print(mayor2)
print(menor2)



for tienda2 in [tienda2]:
  producto_duplicados2 = tienda2.duplicated(subset='Producto')
  tienda_sin_duplicados2 = tienda2[~producto_duplicados2]

plt.plot(tienda_sin_duplicados2["Precio"], tienda_sin_duplicados2['Calificación'], 'ro')
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.title("Relación entre calificación y precio tienda 2")
plt.show()

mayor = tienda2["Precio"].max()
menor = tienda2["Precio"].min()

mayor2 = tienda_sin_duplicados2["Precio"].max()
menor2 = tienda_sin_duplicados2["Precio"].min()

print(mayor)
print(menor)

print(mayor2)
print(menor2)


for tienda3 in [tienda3]:
  producto_duplicados3 = tienda3.duplicated(subset='Producto')
  tienda_sin_duplicados3 = tienda3[~producto_duplicados3]

plt.plot(tienda_sin_duplicados3["Precio"], tienda_sin_duplicados3['Calificación'], 'ro')
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.title("Relación entre calificación y precio tienda 3")
plt.show()

mayor = tienda3["Precio"].max()
menor = tienda3["Precio"].min()

mayor2 = tienda_sin_duplicados3["Precio"].max()
menor2 = tienda_sin_duplicados3["Precio"].min()

print(mayor)
print(menor)

print(mayor2)
print(menor2)


for tienda4 in [tienda4]:
  producto_duplicados4 = tienda4.duplicated(subset='Producto')
  tienda_sin_duplicados4 = tienda4[~producto_duplicados4]

plt.plot(tienda_sin_duplicados4["Precio"], tienda_sin_duplicados4['Calificación'], 'ro')
plt.xlabel("Precio")
plt.ylabel("Calificación")
plt.title("Relación entre calificación y precio tienda 4")
plt.show()

mayor = tienda4["Precio"].max()
menor = tienda4["Precio"].min()

mayor2 = tienda_sin_duplicados4["Precio"].max()
menor2 = tienda_sin_duplicados4["Precio"].min()

print(mayor)
print(menor)

print(mayor2)
print(menor2)
```
